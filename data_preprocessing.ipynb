```python
# Start Washing the Dirty Data
    # For missing data fields, I used Imputer to input the mean.
    # This method IS NOT effective for all missing fields.
dataset = pd.read_csv("/Users/lucyhyde/Desktop/Practicum_Data/org_full_data.csv")

X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 3].values
```

```python
# Imputer Fitting: Quantitative
from sklearn.preprocessing import Imputer
imputer = Imputer(missing_values = np.nan, strategy = ‘mean’, axis = 0)
imputer = imputer.fit(X[:, 1:3])
X[:, 1:3] = imputer.transform(X[:, 1:3])
```

```python
# Categorical Coding and Dummy Variables
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_X = LabelEncoder()
X[:, 0] = labelencoder_X.fit_transform(X[:, 0])
onehotencoder = OneHotEncoder(categorical_features = [0])
X = onehotencoder.fit_transform(X).toarray()

# Dependent Variable (1 = Yes, 0 = No)
labelencoder_y = LabelEncoder()
y = labelencoder_y.fit_transform(y)
print(y)
```
